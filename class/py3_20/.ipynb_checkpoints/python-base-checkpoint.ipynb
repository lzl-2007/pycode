```python
#TempConvert.py
TempStr= input("请输入带有符号的温度值：")
if TempStr[-1] in ['F','f']:
    C = (float(TempStr[0:-1])-32)/1.8
    print("转换后的温度是{:.2f}C".format(C))
elif TempStr[-1] in ['C','c']:
    F = float(TempStr[0:-1])*1.8+32
    print("转换后的温度是{:.2f}F".format(F))
else:
    print("输入格式错误")
```

```python
'''
多行注释
'''
#单行注释


from math import sqrt
from random import random


N = 1000
hits = 0.0
for i in range(0, N, 1):
    x, y= random(), random()
    dist = sqrt(x ** 2 + y ** 2)
    if dist <= 1.0:
        hits += 1
pi=4*(hits/N)
print("Pi的值是{:.2f}".format(pi))

i=0
while i < 5:
    print(i)
    i+=1
for i in range(5):
    print(i)

fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```

```python
def greet(name , age=16, gender="girl"):
    print(f"Hello, {name}, {age} years old {gender}!")
    
greet("Alice")      # 输出: Hello, Alice, 16 years old girl!
greet("Marry",18)   # 输出: Hello, Marry, 18 years old girl!
greet("Bob",gender="boy") #Hello, Bob, 16 years old boy!
```

```python
from functools import reduce

numbers = [1, 2, 3, 4]
squared = list(map(lambda x: x ** 2, numbers))  # 使用 map 和 lambda
print(squared)
filtered = list(filter(lambda x: x % 2 == 0, numbers))  # 使用 filter 和 lambda
print(filtered)
product = reduce(lambda x, y: x if x > y else y, numbers)  # 使用 reduce 和 lambda
print(product)
```

向量化是非常基础的去除for循环的技术，在深度学习实践中，大数据的训练，对代码运行速度的要求比较苛刻，我们都希望更快的得到结果，然后根据训练结果进行模型调参，所以在深度学习领域，向量化是一个关键技巧。

```python
import numpy as np #导入numpy库
import time #导入时间库
a = np.random.rand(1000000)
b = np.random.rand(1000000) #通过round随机得到两个一百万维度的数组a和b
#向量化的版本
tic = time.time() #现在测量一下当前时间 
c = np.dot(a,b)
toc = time.time()
print("%.9f"%c)
print("向量化方法运行时间:" + str(1000*(toc-tic)) +"ms") #打印一下向量化的版本的时间
#继续增加非向量化的版本 
c = 0 
tic = time.time()
for i in range(1000000): 
  c += a[i]*b[i] 
toc = time.time()
print("%.9f"%c)
print("For循环方法运行时间:" + str(1000*(toc-tic)) + "ms")#打印for循环的版本的时间
```
